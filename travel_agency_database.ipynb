### Bir fonksiyon oluşturun

`city` ve `key` olmak üzere iki argüman alan ve şehir için karşılık gelen değeri döndüren `get_key_value_from_city` fonksiyonunu oluşturun.

__Örnek__: `get_key_value_from_city(city=paris_city, key="total_population")` 2102650 döndürecektir.

<u>Not:</u> Eğer bir anahtar (key) yoksa, fonksiyon `"_<key>_ not defined"` döndürmelidir.
    




```python
def get_key_value_from_city(city, key):
    """
    Verilen sözlükten (city) istenen anahtarın (key) değerini döndürür.
    Anahtar yoksa özel bir hata mesajı döner.
    """
    if key in city:
        return city[key]
    else:
        return f"_{key}_ not defined"
```

### Değerleri okuma


Önceki `get_key_value_from_city` fonksiyonunuzu kullanarak iki argüman alan yeni bir fonksiyon oluşturun: `city_name` ve `key`.

Eğer şehir listede yoksa, fonksiyon `"*city* doesn't exist or is not covered"` döndürmelidir.

Fonksiyonunuzu çeşitli örneklerle test edin.

```python
# 1. Adım: İlk fonksiyonumuzu tanımlayalım (veya daha önce tanımladıysan onu kullanır)
def get_key_value_from_city(city, key):
    if key in city:
        return city[key]
    else:
        return f"_{key}_ not defined"

# 2. Adım: Tüm şehir listesinde arama yapan yeni fonksiyon
def get_key_value_from_all_cities(city_name, key):
    # cities listesi içindeki her bir şehir sözlüğünü tek tek kontrol edelim
    for city in cities:
        if city["name"] == city_name:
            # Şehri bulduğumuzda, önceki fonksiyonu kullanarak değeri döndürelim
            return get_key_value_from_city(city, key)
    
    # Döngü biter ve şehir bulunamazsa bu mesajı döndürelim
    return f"*{city_name}* doesn't exist or is not covered"

# --- Testler ---

# Örnek 1: Mevcut bir şehir ve mevcut bir anahtar
print(f"Test 1 (Paris Nüfus): {get_key_value_from_all_cities('Paris', 'total_population')}")

# Örnek 2: Mevcut bir şehir ama olmayan bir anahtar
print(f"Test 2 (Brooklyn Landmark): {get_key_value_from_all_cities('Brooklyn', 'landmark')}")

# Örnek 3: Listede olmayan bir şehir
print(f"Test 3 (İstanbul Kontrol): {get_key_value_from_all_cities('Istanbul', 'total_population')}")

# Örnek 4: Büyük/Küçük harf duyarlılığı testi (Xi’an)
print(f"Test 4 (Xi’an Kıta): {get_key_value_from_all_cities('Xi’an', 'continent')}")
```

### Değer güncelleme


Bazen şehir bilgileri yanlış olabilir veya güncellemeye ihtiyaç duyabilir.

`modify_city_key_value` fonksiyonunu oluşturun; bu fonksiyon mevcut bir anahtarın değerini yeni bir değerle değiştirir. \
Fonksiyon üç argüman alır: `city_name`, `key` ve anahtara atanacak yeni `value`.

Fonksiyon aşağıdaki durumlardan birini belirten bir metin mesajı üretecektir:

- Eğer şehir yoksa veya kapsanmıyorsa: `"*city* doesn't exist or is not covered."`
- Eğer verilen anahtar yoksa: `"*key* doesn't exist."`
- Eğer anahtar başarıyla güncellendiyse: `"*key* has been updated for *city*."`

```python
def modify_city_key_value(city_name, key, new_value):
    # 1. Şehri bulmak için listeyi dönelim
    for city in cities:
        if city["name"] == city_name:
            # 2. Şehir bulundu, şimdi anahtar (key) kontrolü yapalım
            if key in city:
                # Anahtar varsa değeri güncelleyelim
                city[key] = new_value
                return f"*{key}* has been updated for *{city_name}*."
            else:
                # Şehir var ama anahtar yoksa
                return f"*{key}* doesn't exist."
    
    # 3. Döngü bitti ve şehir bulunamadıysa
    return f"*{city_name}* doesn't exist or is not covered."

# --- Test Senaryoları ---

# Senaryo 1: Başarılı Güncelleme (Paris nüfusunu güncelleyelim)
print(modify_city_key_value("Paris", "total_population", 2200000))

# Senaryo 2: Şehir var ama anahtar yanlış/yok
print(modify_city_key_value("Paris", "mayor", "Anne Hidalgo"))

# Senaryo 3: Şehir listede yok
print(modify_city_key_value("Istanbul", "total_population", 15000000))

# Değişikliği kontrol edelim (Paris'in yeni nüfusu)
print(f"Paris Yeni Nüfus: {cities[0]['total_population']}")
```

### Arama motorunu tasarlayın


`get_cities_nearby` adlı bir fonksiyon oluşturun; bu fonksiyon iki argüman alır:
- `customer_location`: [lat, long] formatında koordinatlardan oluşan bir liste


- `radius`: kilometre cinsinden eşik değeri

Ve verilen `customer_location` yakınında bulunan `şehirlerin bir listesini` döndürür.

Eğer hiçbir şehir bulunamazsa, fonksiyon boş bir liste döndürmelidir.

Test etmek için Çin'den Guangyan şehrini ([32.4355, 105.8436]) ve 500 km'lik bir yarıçapı kullanabilirsiniz.

Xi'an ve Chengdu'nun çok uzak olmadığını görmelisiniz...

```python
from geopy import distance

def get_cities_nearby(customer_location, radius):
    """
    Belirli bir konuma radius (km) mesafesindeki şehirleri döndürür.
    """
    nearby_cities = []
    
    for city in cities:
        # Şehrin konum bilgisini alalım [lat, lon]
        city_location = city["location"]
        
        # İki nokta arasındaki mesafeyi hesaplayıp kilometreye çevirelim
        # Mesafe nesnesinden .km özelliğini alarak sayısal (float) bir değer elde ederiz
        d = distance.distance(customer_location, city_location).km
        
        # Eğer mesafe belirlenen yarıçaptan küçük veya eşitse listeye ekle
        if d <= radius:
            nearby_cities.append(city["name"])
            
    return nearby_cities

# --- Test ---
# Guangyan şehri koordinatları
guangyan_location = [32.4355, 105.8436]
esik_degeri = 500 # km

yakin_sehirler = get_cities_nearby(guangyan_location, esik_degeri)

print(f"Guangyan'a {esik_degeri} km mesafedeki şehirler: {yakin_sehirler}")
```
